```python
%matplotlib inline 
# this above line is required to show plots here
# plotting of a linear function (the above line is required to )
import matplotlib.pyplot as plt
import numpy as np # please install numpy and use it

import sys, os
# make the modules available
sys.path.insert(1, os.getcwd() + '/../modules/')

import analytic.channelFlow as cf
H = 1e-3
G = 1e3
mu = 8.9e-4

x = np.linspace(-H/2,H/2,100)
y = cf.poiseuille(x, mu, H, G)
fig = plt.figure(dpi=100)
ax = fig.gca()

ax.plot(x,y);
ax.set_xlim([-H/2,H/2])
ax.set_ylim([0,1.1*max(y)])
ax.set_xlabel('x in m')
ax.set_ylabel('v in m/s')
ax.grid()
```
